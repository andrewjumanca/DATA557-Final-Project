```python
plt.title('Salary by Academic Field')
plt.hist(df.salary[df.field=='Other'], bins=50, alpha=0.7, label='Other');
plt.hist(df.salary[df.field=='Arts'], bins=50, label='Arts');
plt.hist(df.salary[df.field=='Prof'], bins=50, alpha=0.5, label='Prof');
plt.legend()
```

```python
plt.hist(df.salary[df.startyr=='Other'], bins=50, alpha=0.7, label='Other');
plt.hist(df.salary[df.startyr=='Arts'], bins=50, label='Arts');
plt.hist(df.salary[df.startyr=='Prof'], bins=50, alpha=0.5, label='Prof');
plt.legend()
```

```python
plt.figure(figsize=(20,7))
sns.violinplot(data=df, x="year", y="salary", hue="sex")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.title(f'Salary of Males grouped by rank')
sns.violinplot(x=df.year.values[df.sex.values=='M'], y=df.salary.values[df.sex.values=='M'], hue=df['rank'][df.sex.values=='M'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.title('Female violin plot')
sns.violinplot(x=df.year.values[df.sex.values=='F'], y=df.salary.values[df.sex.values=='F'], hue=df['rank'][df.sex.values=='F'])
plt.grid()
plt.show()
```
